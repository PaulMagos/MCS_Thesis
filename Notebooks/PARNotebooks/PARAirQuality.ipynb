```python
from tsl.datasets import MetrLA, PemsBay, AirQuality
import pickle 
import pandas as pd
from os import path
from deepecho import PARModel
dataset = AirQuality(impute_nans=True, small=True)
df3 = dataset.dataframe()[-1000:]
df3.columns = df3.columns.droplevel('channels')
df3.columns = [str(col) for col in df3.columns]

cols = list(df3.columns)
datetime = df3.index
df3 = pd.DataFrame(df3)
df3.columns = cols
df3.insert(0, 'Date', datetime)
```

```python
# Learn a model from the data
if path.exists('pam_airquality.pkl'):
    with open('pam_airquality.pkl', 'rb') as f:
        model = pickle.load(f)
else:
    model = PARModel(epochs=10, cuda=False)
    model.fit(
        data=df3,
        segment_size=1,
        sequence_index='Date'
    )
    with open('pam_airquality.pkl','wb') as f:
        pickle.dump(model,f)
```

```python
data_synth = model.sample(num_entities=1000)
```
